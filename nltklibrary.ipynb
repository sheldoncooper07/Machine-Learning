```
newdata1=[i for i in mydata.split()]
```

```
import matplotlib.pyplot as plt
import nltk

# matplotlib inline   #incase matplotlibv is not supporting by default with pandas or nltk
```

```
[i for i in dir(nltk) if 'Freq' in i] # this will plot data for us
```

```
nlpdata=nltk.FreqDist(newdata1)
```

```
nlpdata.plot(20) #top 20 words
```

```
#we can remove stopwords from newdata1 then plot the graph
from nltk.corpus import stopwords
```

```
#removing stopwords 
removedata=[i for i in newdata1 if i.lower() not in stopwords.words('english')]
```

```
#apply NLTK FreqDist
nlpremove=nltk.FreqDist(removedata)
```

```
nlpremove.plot(20) #top 20 words
```
